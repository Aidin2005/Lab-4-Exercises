```python
# %%
# 📘 Intelligent Data Analysis - Lab 4 Report (Logistics dataset)
# EDA notebook focused on target = "rate"
# Run each cell in order.

# --- Imports and display settings ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import Markdown as md, display
import textwrap

pd.set_option('display.max_columns', None)
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 5)

# Helper for nicely printing small markdown blocks
def show_md(text):
    display(md(text))

# %%
# --- Lab Objective (human, structured, professional) ---
lab_objective = """
### Lab Objective

The main purpose of this lab is to perform an exploratory data analysis (EDA) on a logistics dataset
to understand how different operational factors affect the **rate** (price per load). The dataset contains
details about trucks, trailers, drivers, brokers, pickup and delivery locations, mileage, and stops.

This analysis focuses on identifying patterns, relationships, and potential predictors of `rate`.
In particular, we examine variables such as **loaded miles**, **empty miles**, **stops**, and **broker**
for their relationship with pricing. We also highlight data quality issues and propose next steps
for modeling or operational recommendations.
"""
show_md(lab_objective)

# %%
# --- Load dataset ---
file_path = "logistics_data.csv"  # adjust path if needed
df = pd.read_csv(file_path)
show_md(f"**Dataset loaded:** `{file_path}` — shape: **{df.shape[0]} rows x {df.shape[1]} columns**")
display(df.head(6))

# %%
# --- Quick info & missingness ---
show_md("### Quick data overview")
print("Data types and non-null counts:")
display(df.info())

show_md("#### Missing values per column (top 20)")
missing = df.isnull().sum().sort_values(ascending=False)
display(missing.head(20))

# %%
# --- Identify columns of interest ---
# common columns expected in your dataset (we will test presence)
num_candidates = ['stops', 'empty_mile', 'loaded_miles', 'rate', 'rate_per_mile', 'rate_per_mile']
cat_candidates = ['truck', 'trailer', 'driver_name', 'broker', 'pu_location', 'del_location', 'week']

num_cols = [c for c in num_candidates if c in df.columns]
cat_cols = [c for c in cat_candidates if c in df.columns]

show_md(f"**Numeric columns used:** {num_cols}")
show_md(f"**Categorical columns used:** {cat_cols}")

# Convert numeric-like columns to numeric type if necessary
for c in num_cols:
    df[c] = pd.to_numeric(df[c], errors='coerce')

# %%
# --- 1. Univariate Analysis (numerical) ---
show_md("## 1. Univariate Analysis — Numerical features\nWe first look at single-variable distributions to see ranges, central tendency, spread and outliers.\n")

if len(num_cols) == 0:
    show_md("No numeric columns found among expected candidates. Skip numeric univariate analysis.")
else:
    display(df[num_cols].describe().T.round(3))

    # For each numeric column: histogram + boxplot, with explanation and short insight
    for col in num_cols:
        show_md(f"**What we look for — {col}:** distribution, skewness, outliers, and typical range.")
        fig, axes = plt.subplots(1,2, figsize=(14,4))
        sns.histplot(df[col].dropna(), bins=30, kde=True, ax=axes[0])
        axes[0].set_title(f"Distribution of {col}")
        sns.boxplot(x=df[col].dropna(), ax=axes[1])
        axes[1].set_title(f"Boxplot of {col}")
        plt.tight_layout()
        plt.show()

        # short human explanation after plot
        q = df[col].dropna()
        median = q.median() if not q.empty else np.nan
        iqr = (q.quantile(0.75) - q.quantile(0.25)) if not q.empty else np.nan
        insight = f"- **Observations (short):** median ≈ {median:.2f}; IQR ≈ {iqr:.2f}. "
        if q.empty:
            insight = "- No valid values to display."
        else:
            # quick natural sentence about skew/outliers
            skew = q.skew()
            if abs(skew) < 0.5:
                insight += "Distribution roughly symmetric. "
            elif skew > 0:
                insight += "Right-skewed — a few large values/outliers present. "
            else:
                insight += "Left-skewed — a few unusually small values. "
        show_md(insight)

# %%
# --- 1b. Univariate Analysis (categorical) ---
show_md("## 1b. Univariate Analysis — Categorical features\nWe list top categories to understand dominant trucks/brokers/routes.")

if len(cat_cols) == 0:
    show_md("No categorical columns found among expected candidates. Skip categorical univariate analysis.")
else:
    for col in cat_cols:
        show_md(f"**What we look for — {col}:** top categories and how concentrated the data is.")
        top = df[col].value_counts().head(10)
        display(top)
        # barplot of top 10
        plt.figure(figsize=(10, max(3, 0.3*len(top))))
        sns.barplot(y=top.index.astype(str), x=top.values)
        plt.title(f"Top categories for {col} (top {len(top)})")
        plt.xlabel("Count")
        plt.ylabel(col)
        plt.tight_layout()
        plt.show()

        # human insight
        most = top.index[0] if len(top)>0 else None
        if most is not None:
            show_md(f"- **Observation (short):** `{most}` is the most frequent value for **{col}**, appearing **{top.iloc[0]}** times. This suggests concentration in a few categories.")
        else:
            show_md("- No values to show for this column.")

# %%
# --- 2. Bivariate Analysis (target = rate) ---
show_md("## 2. Bivariate Analysis — How features relate to target `rate`\nWe explore relationships between `rate` and numeric/categorical features to spot drivers of rate differences.")

if 'rate' not in df.columns:
    show_md("**Target column `rate` not found in dataset. Bivariate analysis skipped.**")
else:
    # a) loaded_miles vs rate
    if 'loaded_miles' in df.columns:
        show_md("**What we look for — loaded_miles vs rate:** do longer trips pay more in absolute terms? Are there nonlinearities or clusters?")
        plt.figure(figsize=(9,5))
        sns.scatterplot(x='loaded_miles', y='rate', data=df, alpha=0.7)
        sns.regplot(x='loaded_miles', y='rate', data=df, scatter=False, lowess=True, color='red')
        plt.title("Loaded miles vs Rate")
        plt.xlabel("Loaded miles")
        plt.ylabel("Rate")
        plt.tight_layout()
        plt.show()

        # short human observation
        corr_val = df[['loaded_miles','rate']].dropna().corr().iloc[0,1]
        show_md(f"- **Observation:** correlation between `loaded_miles` and `rate` ≈ **{corr_val:.2f}**. This gives a sense if longer trips are generally more expensive in absolute terms.")

    # b) rate_per_mile vs rate (effectiveness)
    if 'rate_per_mile' in df.columns:
        show_md("**What we look for — rate_per_mile vs rate:** see whether higher per-mile rates correspond to higher total rates; spot lanes with high per-mile pay.")
        plt.figure(figsize=(9,5))
        sns.scatterplot(x='rate_per_mile', y='rate', data=df, alpha=0.7)
        sns.regplot(x='rate_per_mile', y='rate', data=df, scatter=False, lowess=True, color='red')
        plt.title("Rate per Mile vs Rate")
        plt.xlabel("Rate per mile")
        plt.ylabel("Rate")
        plt.tight_layout()
        plt.show()

        corr_val2 = df[['rate_per_mile','rate']].dropna().corr().iloc[0,1]
        show_md(f"- **Observation:** correlation between `rate_per_mile` and `rate` ≈ **{corr_val2:.2f}**. This tells if higher per-mile price also means higher total pay (often yes, but check loaded miles effect).")

    # c) stops vs rate (boxplot)
    if 'stops' in df.columns:
        show_md("**What we look for — stops vs rate:** Does adding stops change pricing? (more stops may reduce net efficiency)")
        plt.figure(figsize=(9,5))
        sns.boxplot(x='stops', y='rate', data=df)
        plt.title("Stops vs Rate")
        plt.xlabel("Stops")
        plt.ylabel("Rate")
        plt.tight_layout()
        plt.show()
        show_md("- **Observation:** Compare medians across stop counts — significant shifts indicate pricing adjustments for complexity.")

    # d) broker vs rate (top brokers)
    if 'broker' in df.columns:
        show_md("**What we look for — broker vs rate:** which brokers pay better on average? We show top 10 by mean rate.")
        broker_stats = df.groupby('broker')['rate'].agg(['count','mean','median']).sort_values('mean', ascending=False).head(10)
        display(broker_stats.round(2))

        # horizontal bar for mean rate
        plt.figure(figsize=(10,6))
        sns.barplot(x=broker_stats['mean'].values, y=broker_stats.index.astype(str))
        plt.xlabel("Mean Rate")
        plt.title("Top 10 Brokers by Mean Rate")
        plt.tight_layout()
        plt.show()

        show_md("- **Observation:** Brokers at the top likely represent lanes or customers that pay premium rates. Check their lane patterns and frequency.")

# %%
# --- 3. Correlation Analysis (numeric) ---
show_md("## 3. Correlation analysis\nWe compute correlations between numeric features to understand linear relationships and possible predictors for `rate`.")

if len(num_cols) < 2:
    show_md("Not enough numeric columns for a meaningful correlation matrix.")
else:
    corr = df[num_cols].corr()
    plt.figure(figsize=(8,6))
    sns.heatmap(corr, annot=True, fmt=".2f", cmap="vlag", center=0)
    plt.title("Correlation matrix (selected numeric features)")
    plt.tight_layout()
    plt.show()

    # human-readable summary
    # highlight strong correlations with rate if present
    if 'rate' in corr.index:
        related = corr['rate'].drop('rate').abs().sort_values(ascending=False)
        top_rel = related.head(5)
        summary = "- **Top numeric correlations with `rate`:**\n"
        for ix, val in top_rel.items():
            summary += f"  - `{ix}`: {corr.loc[ix,'rate']:.2f}\n"
        show_md(summary)

# %%
# --- 4. Route / Week Aggregations (if available) ---
show_md("## 4. Aggregations and trends\nWe aggregate by week and by driver/broker to see higher-level patterns.")

# weekly trend
if 'week' in df.columns:
    show_md("**What we look for — weekly trend in `rate`:** do rates vary across weeks? Look for seasonality or peaks.")
    try:
        weekly = df.groupby('week')['rate'].mean().sort_index()
        plt.figure(figsize=(12,4))
        sns.lineplot(x=weekly.index.astype(str), y=weekly.values, marker='o')
        plt.xticks(rotation=45)
        plt.ylabel("Average Rate")
        plt.title("Average Rate by Week")
        plt.tight_layout()
        plt.show()
        show_md("- **Observation:** Look for weeks with spikes/dips — could indicate special events/market changes.")
    except Exception as e:
        show_md(f"- Could not plot weekly trend: {e}")

# top drivers by average rate (if driver present)
if 'driver_name' in df.columns:
    show_md("**What we look for — drivers with highest average rate:** might indicate preferred drivers / lanes.")
    driver_stats = df.groupby('driver_name')['rate'].agg(['count','mean']).sort_values('mean', ascending=False).head(10)
    display(driver_stats.round(2))
    plt.figure(figsize=(10,6))
    sns.barplot(x=driver_stats['mean'].values, y=driver_stats.index.astype(str))
    plt.title("Top 10 Drivers by Average Rate")
    plt.xlabel("Mean Rate")
    plt.tight_layout()
    plt.show()
    show_md("- **Observation:** Drivers at top may work premium lanes or get preferred customers; check counts to ensure it's not single high-value outlier loads.")

# %%
# --- 5. Draft Data Dictionary (sample) ---
show_md("## 5. Draft data dictionary (sample entries)\nQuick look at types, non-null counts, and sample values.")
dd = []
for col in df.columns:
    sample_vals = df[col].dropna().astype(str).unique()[:3].tolist()
    dd.append({
        "column": col,
        "dtype": str(df[col].dtype),
        "non_null": int(df[col].notnull().sum()),
        "sample_values": sample_vals
    })
dd_df = pd.DataFrame(dd)
display(dd_df.head(50))

# %%
# --- 6. 10 Key Insights (human, short) ---
show_md("## 6. 10 Key Insights — concise and human")

insights = []

# populate insights from observed data (examples and conditional)
if 'loaded_miles' in df.columns:
    insights.append(f"Median loaded miles ≈ {df['loaded_miles'].median():.1f}.")
if 'rate_per_mile' in df.columns:
    insights.append(f"Average rate per mile ≈ {df['rate_per_mile'].mean():.2f}.")
if 'rate' in df.columns:
    insights.append(f"Average rate per load ≈ {df['rate'].mean():.2f}.")
if 'stops' in df.columns:
    top_stop = df['stops'].value_counts().idxmax()
    insights.append(f"Most loads have {top_stop} stop(s) — typical pickup+delivery pattern.")
if 'broker' in df.columns:
    top_broker = df['broker'].value_counts().index[0]
    insights.append(f"Data concentrated in a few brokers; top broker is `{top_broker}`.")
# general data quality observations
missing_cols = missing[missing>0]
if not missing_cols.empty:
    insights.append(f"Missing values present in: {', '.join(missing_cols.index[:5].tolist())} (handle before modeling).")
# potential outliers
if 'rate' in df.columns:
    high_rate_count = df[df['rate'] > df['rate'].quantile(0.95)].shape[0]
    insights.append(f"There are {high_rate_count} loads with `rate` in the top 5% — investigate as possible outliers or premium lanes.")
# filler to reach 10 items
extra = [
    "Rate shows moderate correlation with loaded miles and rate_per_mile (see correlation matrix).",
    "Some drivers/brokers consistently appear with higher-than-average rates — good candidates for further analysis.",
    "Empty miles vary widely; high empty miles can hurt profitability and may explain low rate_per_mile.",
    "Next steps should include feature engineering: lane (PU->DEL), driver performance, and time-of-week effects."
]
for e in extra:
    if len(insights) < 10:
        insights.append(e)

# print as a numbered list
for i, it in enumerate(insights[:10], 1):
    show_md(f"**{i}.** {it}")

# %%
# --- 7. Conclusion & Next steps (short, human) ---
concl_md = """
## Conclusion

This exploratory analysis highlights several clear signals in the logistics dataset:

- `rate` varies with both **distance** (`loaded_miles`) and **price per mile** (`rate_per_mile`), but there is meaningful variability by **broker** and **driver**.
- Data quality issues exist (missing pickup/delivery dates and some zero/NaN values); those must be cleaned before modeling.
- There are a few high-value loads that look like outliers — investigate whether they are legitimate premium lanes or data errors.

**Next steps:** data cleaning (impute or drop missing dates), create lane features (PU->DEL), compute driver/broker aggregates, engineer time features, then build predictive models for `rate` or optimize routing/pricing.

"""
show_md(concl_md)

# %%
# Optional: Save aggregated numeric summary to CSV (useful)
num_summary_path = "numeric_summary_logistics.csv"
if len(num_cols):
    df[num_cols].describe().to_csv(num_summary_path)
    show_md(f"Saved numeric summary to `{num_summary_path}`")

# %%
# Optional: Export all plotted matplotlib figures into a single PDF (uncomment to use)
# from matplotlib.backends.backend_pdf import PdfPages
# pdf_out = "EDA_report_plots.pdf"
# with PdfPages(pdf_out) as pdf:
#     # This example just re-creates some key plots and saves them; adapt as needed
#     if 'loaded_miles' in df.columns and 'rate' in df.columns:
#         fig = plt.figure()
#         sns.scatterplot(x='loaded_miles', y='rate', data=df, alpha=0.7)
#         plt.title("Loaded miles vs Rate")
#         pdf.savefig(); plt.close()
#     if 'rate_per_mile' in df.columns:
#         fig = plt.figure()
#         sns.scatterplot(x='rate_per_mile', y='rate', data=df, alpha=0.7)
#         plt.title("Rate per Mile vs Rate")
#         pdf.savefig(); plt.close()
#     # add other figures as needed
# show_md(f"Saved PDF report (plots only) to `{pdf_out}`")
```
